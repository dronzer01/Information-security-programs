```python
def redundant(m):
    r=0
    while(2**r<r+m+1):
        r=r+1
    return r
def pos_redundant(data,m,r):
    j=0
    k=0
    res=''
    for i in range(1,m+r+1):
        if i==2**j:
            res=res+'0'
            j=j+1
        else:
            res=res+data[k]
            k=k+1
    return res
def parity_value_sender(res,m,r):
    lst=list(res)
    p=0
    for i in range(0,r):
        pos=2**i
        s=pos-1
        p=0
        while(s<(m+r)):
            j=s
            while(j<pos+s and j<m+r):
                if lst[j]=="1":
                    p=p+1
                j=j+1
            s=s+2*pos
        if(p%2==0):
            lst[pos-1]="0"
        else:
            lst[pos-1]="1"
    res=''.join(lst)
    return res

def parity_value_receiver(res,m,r):
    lst=list(res)
    bit=0
    for i in range(0,r):
        pos=2**i
        s=pos-1
        p=0
        while(s<(m+r)):
            j=s
            while(j<pos+s and j<m+r):
                if lst[j]=="1":
                    p=p+1
                j=j+1
            s=s+2*pos
        if(p%2!=0):
            bit=bit+pos
    return bit

print("----------ERROR CORRECTION CODE----------")
#Sender's side
data1=input("Enter the msg")
m=len(data1)
r=redundant(m)
print("No. of Redundant Bits=",r)
pres=pos_redundant(data1,m,r)
print("Redundant Data:",pres)
res1=parity_value_sender(pres,m,r)
print("Redundant Data with Parity Value:",res1)

#Receiver's Side
ans=(input("Do you want to Introduce an Error? 'y' or 'n': "))
if ans=='y':
    data2=input("Enter the wrong data value: ")
else:
    data2=res1

res2=parity_value_receiver(data2,m,r)

if(res2!=0):
    print("Error in Message: Error at position:",res2)
    
else:
    print("No Error: Received Data with Parity Value is:",data2)
    j=0
    msg=''
    for i in range(0,len(data2)):
        pos=2**j
        if(i!=(pos-1)):
            msg=msg+data2[i]
        else:
            j=j+1
    print("Data Value of Message Received:", msg)
```

```python
#!/usr/bin/env python
# coding: utf-8

# In[7]:


# Python program to implement Autokey Cipher


# Dictionary to lookup the index of alphabets
dict1 = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4,
         'F': 5, 'G': 6, 'H': 7, 'I': 8, 'J': 9,
         'K': 10, 'L': 11, 'M': 12, 'N': 13, 'O': 14,
         'P': 15, 'Q': 16, 'R': 17, 'S': 18, 'T': 19,
         'U': 20, 'V': 21, 'W': 22, 'X': 23, 'Y': 24, 'Z': 25}

dict2 = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E',
         5: 'F', 6: 'G', 7: 'H', 8: 'I', 9: 'J',
         10: 'K', 11: 'L', 12: 'M', 13: 'N', 14: 'O',
         15: 'P', 16: 'Q', 17: 'R', 18: 'S', 19: 'T',
         20: 'U', 21: 'V', 22: 'W', 23: 'X', 24: 'Y', 25: 'Z'}


# This function generates the key
def generate_key(message, key):
    i = 0
    while True:
        if len(key) == len(message):
            break
        if message[i] == ' ':
            i += 1
        else:
            key += message[i]
            i += 1
    return key


# This function returns the encrypted text
# generated with the help of the key
def cipherText(message, key_new):
    cipher_text = ''
    i = 0
    for letter in message:
        if letter == ' ':
            cipher_text += ' '
        else:
            x = (dict1[letter]+dict1[key_new[i]]) % 26
            i += 1
            cipher_text += dict2[x]
    return cipher_text


# This function decrypts the encrypted text
# and returns the original text
def originalText(cipher_text, key_new):
    or_txt = ''
    i = 0
    for letter in cipher_text:
        if letter == ' ':
            or_txt += ' '
        else:
            x = (dict1[letter]-dict1[key_new[i]]+26) % 26
            i += 1
            or_txt += dict2[x]
    return or_txt


def main():
    message =input('Enter the message')
    key =input('Enter the key')
    message=message.upper()
    key=key.upper()
    key_new = generate_key(message, key)
    cipher_text = cipherText(message, key_new)
    original_text = originalText(cipher_text, key_new)
    print("Encrypted Text =", cipher_text)
    print("Original Text =", original_text)


# Executes the main function
if __name__ == '__main__':
    main()
```
